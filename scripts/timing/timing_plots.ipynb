```python
import matplotlib.pyplot as plt
import numpy as np
from pandas import DataFrame as df

import json
```

```python
timing_data = json.load(open("gpu_test_dur_2.0_dt_5.0_iters_5.json" ,"r"))
```

```python
time_df = df(timing_data)
```

```python
plt.hist(time_df['fd_time'], density=True)
plt.xlabel('time [s]')
plt.show()
```

```python
plt.hist(time_df['td_time'], density=True)
plt.xlabel('time [s]')
plt.show()
```

```python
tmp = time_df[['mass_1', 'p0', 'td_time','fd_time']].to_numpy()
plt.scatter(tmp.T[0], tmp.T[2], c='tab:blue', label='td')
plt.scatter(tmp.T[0], tmp.T[3], c='tab:orange', label='fd')
plt.xlabel(r'$M$')
plt.ylabel('time [s]')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.ylim(0.1,0.5)
plt.show()
```

```python
tmp = time_df[['mass_1', 'p0', 'e0', 'td_time','fd_time']].to_numpy()
plt.scatter(tmp.T[2], tmp.T[3], c='tab:blue', label='td')
plt.scatter(tmp.T[2], tmp.T[4], c='tab:orange', label='fd')
plt.xlabel(r'$e_0$')
plt.ylabel('time [s]')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.show()
```
